 # Implement a single-layer perceptron and a two-layer feedforward network using NumPy; manually compute forward propagation and loss for a small dataset.

#### Importing the Libraries

```python


import numpy as np

```

#### Activation Functions & Loss

```python

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def relu(z):
    return np.maximum(0, z)

def binary_cross_entropy(y_true, y_pred):
    eps = 1e-15
    y_pred = np.clip(y_pred, eps, 1 - eps)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

```

#### Single Layer Perceptron

```python
class SingleLayerPerceptron:
    def __init__(self, weights, bias):
        self.W = np.array(weights, dtype=np.float64)
        self.b = np.array(bias, dtype=np.float64)

    def forward(self, X):
        self.z = np.dot(X, self.W) + self.b
        self.a = sigmoid(self.z)
        return self.a

```

#### Multi Layer Perceptron

```python

class TwoLayerNN:
    def __init__(self, W1, b1, W2, b2):
        self.W1 = np.array(W1, dtype=np.float64)
        self.b1 = np.array(b1, dtype=np.float64)
        self.W2 = np.array(W2, dtype=np.float64)
        self.b2 = np.array(b2, dtype=np.float64)

    def forward(self, X):

        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = relu(self.z1)


        self.z2 = np.dot(self.a1, self.W2) + self.b2
        self.a2 = sigmoid(self.z2)
        return self.a2

```

#### Main Function

```python

if __name__ == "__main__":
    # Input Dataset (2 samples, 2 features)
    X = np.array([[1.0, 2.0],
                  [-1.0, 0.5]])

    # Ground Truth Target Labels
    y = np.array([[1.0],
                  [0.0]])

    print("============================================")
    print("      NEURAL NETWORK EXPERIMENT OUTPUT      ")
    print("============================================\n")

    # 1. Single-Layer Perceptron Verification
    slp = SingleLayerPerceptron(weights=[[0.5], [-0.25]], bias=[0.1])
    y_pred_slp = slp.forward(X)
    loss_slp = binary_cross_entropy(y, y_pred_slp)

    print("[1] Single-Layer Perceptron Results:")
    print(f" Predicted Probabilities: {y_pred_slp.ravel()}")
    print(f" Computed BCE Loss:       {loss_slp:.6f}\n")

    # 2. Two-Layer Neural Network Verification
    W1 = [[0.5, -0.5], [1.0, 0.0]]
    b1 = [0.0, 0.2]
    W2 = [[0.8], [-1.0]]
    b2 = [-0.1]

    mlp = TwoLayerNN(W1, b1, W2, b2)
    y_pred_mlp = mlp.forward(X)
    loss_mlp = binary_cross_entropy(y, y_pred_mlp)

    print("[2] Two-Layer Feedforward Network Results:")
    print(f" Predicted Probabilities: {y_pred_mlp.ravel()}")
    print(f" Computed BCE Loss:       {loss_mlp:.6f}")
    print("============================================")
```
